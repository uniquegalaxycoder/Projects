check the null value in percentage format

```python
null_value = data.isnull().sum()/data.shape[0] * 100
null_value
```

```python
drop_column = null_value[null_value > 10].keys()
```

```python
drop_column
```

```python
dataset = data.drop(columns = drop_column)
```

> after drop null values column check shape

```python
dataset.shape
```

```python
plt.figure(figsize = (25,25))
sns.heatmap(dataset.isnull())
```
