# players based on age group

```python
df.head(1)
```

```python
print(df['Age'].min())
print(df['Age'].mean())
print(df['Age'].max())
```

```python
under_20 =df[ df['Age'] <= 20 ]
age20_25 = df[(df['Age'] >20) & (df['Age'] <=25)]
age25_30 = df[(df['Age'] >25) & (df['Age'] <=30)]
above_30 = df[df['Age'] > 30]
```

```python
data1 = np.array([under_20['Name'].count(), age20_25['Name'].count(), age25_30['Name'].count(), above_30['Name'].count() ])
```

```python
plt.figure(figsize=(13,6))
lables1 = ["<=20", ">20 & <=25", ">25 & <=30", ">30"]
plt.title("Total Player with Age Group", fontsize= 18)
plt.pie(data1, labels= lables1, autopct='%.1f%%', colors = sns.color_palette('Set2'))
plt.show()
```

# Total assistance from each club
